# 🚧 Tokenización
Ahora vamos a pasar el texto del objeto pdf a una lista de palabras. Vamos a tokenizar el texto de cada página mientras la leemos. Vamos a comparar diferentes métodos de tokenizacion y ver cual es el que mejor funciona para nosotros.

- wordpunct_tokenize de NLTK nos parece conveniente al separa los elementos de puntuacion de las palabras de las palabras en las que aparece

- Como estamos tokenizando un texto de tamaño considerable vamos a hacer un tokenizador "manual" mucho más simple utilizando expresiones regulares, cuyo comportamiento es completamente predecible por nosotros y compararemos las diferencias entre los dos conjuntos de tokenes.

- No debemos ignorar los beneficios de utilizar un algoritmo de tokenizado como PUNKT ahora que estamos trabajando con un corpus en inglés

```python
# Preparación del tokenizer manual
import re
pattern = re.compile(r'[^a-zA-Z0-9]') #Caracteres de interés en inglés
```

```python
from nltk.tokenize import wordpunct_tokenize
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk.data
nltk.download('punkt')
```

```python
tokens = []
tokensManuales = []
tokensPunkt = []
# Recorrer todas las páginas del txt
page_text = []

with open('../Corpus/Txt/TheFinalEmpire.txt', 'r', encoding='utf-8') as f:
    page_text.extend(f.readlines())

with open('../Corpus/Txt/TheWellOfAscension.txt', 'r', encoding='utf-8') as f:
    page_text.extend(f.readlines())

# Unir las líneas en una sola cadena sin caracteres '\n'
page_text = ''.join(page_text)
print(page_text[:100])
## -- Tokenizado NLTK --##
pageTokens = wordpunct_tokenize(page_text)
tokens.extend(pageTokens)

## -- Tokenizado manual (regex) --##
# Dividimos el texto en palabras (asumiendo que las palabras están separadas por espacios)
words = page_text.split()
# Iterar a través de las palabras y eliminar caracteres especiales
cleaned_words = [token for word in words for token in pattern.sub(' ', word).split()]
tokensManuales.extend(cleaned_words)

## -- PUNKT tokenizer --##
sentences = sent_tokenize(page_text)
# Tokenizamos cada oración en palabras
for sentence in sentences:
    words = word_tokenize(sentence)
    tokensPunkt.extend(words)
```

Observamos los resultados:

```python
print(tokens[:15])
print(f"Tamaño: {len(set(tokens))}")
print(tokensManuales[:15])
print(f"Tamaño: {len(set(tokensManuales))}")
print(tokensPunkt[:15])
print(f"Tamaño: {len(set(tokensPunkt))}")
```

```python
setNLTK = set(tokens)
setManual = set(tokensManuales)
setPunkt = set(tokensPunkt)

NLTKnotManual = setNLTK - setManual
manualNotNLTK = setManual - setNLTK

NLTKnotPunkt = list(setNLTK - setPunkt)
PunktNotNLTK = list(setPunkt - setNLTK)

NLTKnotManual_list = list(NLTKnotManual)
manualNotNLTK_list = list(manualNotNLTK)



print(f"En NLTK pero no en manual:\n{NLTKnotManual_list}\nTamaño: {len(NLTKnotManual_list)}")
print(f"En manual pero no en NLTK:\n{manualNotNLTK_list}\nTamaño: {len(manualNotNLTK_list)}")

print(f"En NLTK pero no en PUNK:\n{NLTKnotPunkt}\nTamaño: {len(NLTKnotPunkt)}")
print(f"En PUNKT pero no en NLTK:\n{PunktNotNLTK}\nTamaño: {len(PunktNotNLTK)}")
```

Tambien podemos ver los tokens mas frecuentes de cada conjunto de tokens

```python
def getDiccionarioAppearences(listaTokens):
    palabrasDic = {}
    for palabra in listaTokens:
        if palabra in palabrasDic:
            palabrasDic[palabra] +=1
        else:
            palabrasDic[palabra] = 1
    return palabrasDic

def getPalabrasMasFrecuentes(palabrasDic, k):
    palabrasDic = dict(sorted(palabrasDic.items(), key = lambda x:-x[1]))
    return list(palabrasDic)[:k]
```

```python
dictTokens = getDiccionarioAppearences(tokens)
print(f"Palabras mas frecuentes NLTK: {list(getPalabrasMasFrecuentes(dictTokens, 10))}")
dictManuales = getDiccionarioAppearences(tokensManuales)
print(f"Palabras mas frecuentes regex: {list(getPalabrasMasFrecuentes(dictManuales, 10))}")
dictPunkt = getDiccionarioAppearences(tokensPunkt)
print(f"Palabras mas frecuentes PUNKT: {list(getPalabrasMasFrecuentes(dictPunkt, 10))}")
```

# 📈 TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def load_file(filename):
    with open(filename, 'r') as file:
      loaded_file = file.read()
      return loaded_file


book_1 = load_file("../Corpus/Txt/TheFinalEmpire.txt")
book_2 = load_file("../Corpus/Txt/TheHeroOfAges.txt")

corpus = [book_1, book_2]
vectorizer = TfidfVectorizer(min_df=2, stop_words='english')
model = vectorizer.fit_transform(corpus)

# Compute the cosine similarities
similarities = cosine_similarity(model)

doc_names = ["The Final Empire", "The Hero of Ages"]

# Manually printing the cosine similarities
print(f"Cosine similarity between {doc_names[0]} and {doc_names[1]}: {similarities[0][1]:.2f}")
```

```python
import numpy as np

# Transforma el modelo TF-IDF a un arreglo
tfidf_array = model.toarray()

# Obtiene las palabras (features) del vectorizador
feature_names = vectorizer.get_feature_names_out()

# Función para obtener las palabras más relevantes
def get_top_words(tfidf_vector, feature_names, top_n=10):
    """Retorna las top_n palabras con el mayor valor TF-IDF."""
    sorted_nzs = np.argsort(tfidf_vector)[-top_n:][::-1]
    return [(feature_names[i], tfidf_vector[i]) for i in sorted_nzs]

# Obtiene las palabras más relevantes para cada documento
for idx, doc in enumerate(corpus):
    print(f"Documento: {doc_names[idx]}")
    top_words = get_top_words(tfidf_array[idx], feature_names)
    for word, score in top_words:
        print(f"{word}: {score:.4f}")
    print("--------")
```
