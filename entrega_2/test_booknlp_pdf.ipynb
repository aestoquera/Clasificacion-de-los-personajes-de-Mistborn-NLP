```python
import PyPDF2
#Necesario para cargar PDFs. Es una libreria que hay que instalarla a parte

#Creamos el objeto PDF
pdfFile = open("../Corpus/The_Well_of_Ascension_-_Brandon_Sanderson.pdf", "rb")
#Creamos el objeto lector
pdfReader = PyPDF2.PdfReader(pdfFile)
```

```python
pdfReader.pages[450].extract_text()
```

```python
# Creamos txt
text = ""
page_num = 1
ignore = list(range(1,13))
for page in range(902,911):
    ignore.append(page)


for page in pdfReader.pages:
    if page_num in ignore:
        continue

    page_text = page.extract_text().encode('utf-8', 'ignore').decode('utf-8')

    page_text = page_text.replace("OceanofPDF.com", "")

    page_num += 1
    text += page_text 

with open('../Corpus/TheWellOfAscension.txt', 'w', encoding='utf-8') as text_file:
    text_file.write(text)

# He borrado algunas lineas a mano
```

```python
from booknlp.booknlp import BookNLP

model_params={
		"pipeline":"entity,quote,supersense,event,coref",
		"model":"big",
        "entity_model_path":r'C:\Users\theal\booknlps\entities_google\bert_uncased_L-6_H-768_A-12',
        "coref_model_path":r"C:\Users\theal\booknlps\coref_google\bert_uncased_L-12_H-768_A-12",
        "quote_attribution_model_path":r"C:\Users\theal\booknlps\speaker_google\bert_uncased_L-12_H-768_A-12",
	}

booknlp=BookNLP("en", model_params)

# Input file to process
input_file="../Corpus/Mistborn_Trilogy.txt"

# Output directory to store resulting files in
output_directory="../BookNLP/"

# File within this directory will be named ${book_id}.entities, ${book_id}.tokens, etc.
book_id="mistborn"

booknlp.process(input_file, output_directory, book_id)
```
