#  Clases auxiliares

```python
class EpochLogger(CallbackAny2Vec):
    '''Callback to log information about training'''

    def __init__(self):
        self.epoch = 0

    def on_epoch_end(self, model):
        loss = model.get_latest_training_loss()
        if self.epoch == 0:
            print('Loss after epoch {}: {}'.format(self.epoch, loss))
        else:
            print('Loss after epoch {}: {}'.format(self.epoch, loss- self.loss_previous_step))
        self.epoch += 1
        self.loss_previous_step = loss


#Tienes que proporcionar a esta clase una lista de rutas con txts.
class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __init__(self, paths):
        self.paths = paths

    def __iter__(self):
        len_lines = 0
        for path in self.paths:
            with open(path, 'r') as file:
                for line in file:
                    len_lines += 1
                    yield utils.simple_preprocess(line)
```

```python
from nltk.tokenize import wordpunct_tokenize
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS as sklearn_stop_words
nltk.download('stopwords')
nltk.download('punkt')

def tokens_from_text(raw_text):
    tmp_tokens = [word for sentence in sent_tokenize(raw_text.lower()) for word in word_tokenize(sentence) if word.isalpha() and len(word) > 1]
    stopwords = nltk.corpus.stopwords.words('english')
    tmp_tokens_no_stopwords = [x for x in tmp_tokens if x not in stopwords]
    return tmp_tokens_no_stopwords
```

```python
elmo = Elmo(options_file, weight_file, 1, dropout=0)
```

```python
import torch.nn.functional as F #F.cosine_similarity

def get_elmo_embeddings(sentences):
    character_ids = batch_to_ids(sentences)
    embeddings = elmo(character_ids)
    return embeddings['elmo_representations'][0]



#TODO: INSERTAR oraciones con palabras tokenizadas en este array.
sentences = ["Extroverted".split(), "Introverted".split(), "Thinker".split()]
embeddings = get_elmo_embeddings(sentences)

print(embeddings.shape)
print(embeddings[0].shape)

import torch.nn.functional as F
embedding1 = embeddings[0]
embedding2 = embeddings[1]

print(embedding2)

def average_embedding(sentence_embedding):
    return sentence_embedding.mean(dim=0)

similarity = F.cosine_similarity(average_embedding(embedding1), average_embedding(embedding2), dim=0)
print(f"La similitud del coseno entre las dos oraciones es: {similarity.item()}")
```
